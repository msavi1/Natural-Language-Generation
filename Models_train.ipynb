```python
import matplotlib.pyplot as plt
import matplotlib.ticker as plticker
from matplotlib.pyplot import figure
import numpy as np
```

**MODEL 1**

```python
Train_accuracy = [0, 54.52, 62.87, 58.58, 63.36, 68.91, 68.14, 71.57, 72.56, 69.48, 73.79]
Validation_accuracy = [0, 56.5282, 62.8837, 62.1406, 65.6744, 66.3765, 67.5393, 68.7818, 69.0089, 69.1849, 69.5415]
Train_steps = [0, 10000, 20000, 30000, 40000, 50000, 60000, 70000, 80000, 90000, 100000]

plt.figure(figsize=(9, 5))
plt.plot(Train_steps, Train_accuracy, label="Train accuracy")
plt.plot(Train_steps, Validation_accuracy, label="Validation accuracy")

plt.ylabel('Accuracy')
plt.xlabel('Train steps')
plt.legend()

plt.ylim(0, 100)

plt.xticks(np.arange(0, 100000+1, 10000))
plt.yticks(np.arange(0, 100+1, 10))
plt.title('Model1: train and validation accuracy vs. number of training iterations', pad=10)

#plt.savefig('Model1_train.png')
plt.show()
```
